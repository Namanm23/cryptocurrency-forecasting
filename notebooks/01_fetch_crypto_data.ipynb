# Cryptocurrency Data Collection
This notebook fetches historical OHLCV (Open, High, Low, Close, Volume) data
for 10 major cryptocurrencies from Yahoo Finance using the `yfinance` library
and saves them as individual CSV files in the `data/raw/` directory.

## Import Libraries

```python
import yfinance as yf
import pandas as pd
import os
```

## Define Cryptocurrencies
Mapping of file names to their Yahoo Finance ticker symbols.

```python
cryptos = {
    "ADA": "ADA-USD",
    "BNB": "BNB-USD",
    "BTC": "BTC-USD",
    "DOGE": "DOGE-USD",
    "DOT": "DOT-USD",
    "ETH": "ETH-USD",
    "LINK": "LINK-USD",
    "LTC": "LTC-USD",
    "SOL": "SOL-USD",
    "XRP": "XRP-USD",
}
```

## Configuration
Define output directory and date range for data collection.

```python
output_dir = "../data/raw"
os.makedirs(output_dir, exist_ok=True)

START = "2021-01-01"
END   = "2025-12-31"
```

## Download & Save Data
For each cryptocurrency:
- Download OHLCV data from Yahoo Finance
- Flatten multi-level columns
- Keep only relevant columns: Date, Open, High, Low, Close, Adj Close, Volume
- Drop Adj Close (same as Close for crypto — no dividends/splits)
- Save as individual CSV files in `data/raw/`

```python
for name, ticker in cryptos.items():
    print(f"Downloading {name} ({ticker})...")

    try:
        df = yf.download(ticker, start=START, end=END, auto_adjust=False, progress=False)

        if df.empty:
            print(f"No data found for {ticker}, skipping.")
            continue

        df.columns = [col[0] if isinstance(col, tuple) else col for col in df.columns]
        df = df.reset_index()
        df = df[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
        df["Date"] = pd.to_datetime(df["Date"]).dt.strftime("%b %d, %Y")

        output_path = os.path.join(output_dir, f"{name}.csv")
        df.to_csv(output_path, index=False)
        print(f"  ✅ Saved: {output_path}  |  Rows: {len(df)}")

    except Exception as e:
        print(f"Error for {ticker}: {e}")

print("\nDone! All files saved in:", output_dir)
```

## Verify Saved Files
Check all files are saved correctly in `data/raw/` with expected row counts.

```python
print("Files saved in data/raw/:")
print("-" * 40)
for file in sorted(os.listdir(output_dir)):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(output_dir, file))
        print(f"  {file:<12} | Rows: {len(df)} | Columns: {list(df.columns)}")
```
